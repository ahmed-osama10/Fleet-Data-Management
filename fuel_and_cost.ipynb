```python
import pandas as pd
df=pd.read_excel('fuel and cost.xlsx')
df.head()
```

```python
df = df.drop(columns=['بطاريه', 'كاوتش'])
df.head()
```

```python
df.info()
```

```python
df = df.drop(df.tail(3).index)
df = df[:-3]
df.tail()
```

```python
df['الاجمالى لمصروف كل سياره'] = df['الاجمالى لمصروف كل سياره'].astype(int)
df['زيوت يناير'] = df['زيوت يناير'].astype(int)
df['صيانه يناير'] = df['صيانه يناير'].astype(int)
df['وقود يناير'] = df['وقود يناير'].astype(int)
df['كيلومتر يناير'] = df['كيلومتر يناير'].astype(int)
df['رقم السيارة ']=df['رقم السيارة '].astype(int)
df.info()
```

```python
df.tail()
```

```python
df.to_excel('cleaned fuel and cost.xlsx', index=False)
```
