```python
# Step 1: Switch day and month if needed
def switch_day_month(date):
    if date.month != 1:
        return date.replace(day=date.month, month=1)
    return date

# Step 2: Apply the transformations to the "Date" column
df["Date"] = pd.to_datetime(df["Date"])
df["Date"] = df["Date"].apply(switch_day_month).dt.strftime("%Y-%m-%d")

# Display the modified dataframe
print(df["Date"])
```
